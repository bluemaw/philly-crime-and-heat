### A discussion of how your analysis is limited and could be improved:
---

#### Crime Data Limitations
A heavy limitation regarding the crime data we have is already inherent in the data we are using. This data is specifically crime incident-reporting data. From our perspective, there are three major limitations from this:

1. A crime has to be noticed to be reported, not all crimes are reported as some are considered "victimless" crimes. For these crimes, there would be no reporter. Hence, our data would not include it. 
2. Time of incident-report. A crime is reported when it is noticed. For this reason the likelihood of late night crimes being under-reported is high. Two example scenarios would be: a burglary of a business or a theft from a vehicle. Those crimes would likely be reported in the morning when they are noticed, but the crime itself would have been committed the night prior.
3. Crime reporting rate versus actual crime rate. Our crime data does **not** include conviction data. A crime may be reported that did not occur; just as conversely, a crime may occur that does not get reported. There are likely many cases where we will encounter a crime report (data-point) and it does not accurately represent that incident.

#### Temperature Data Limitations
1. Not accurate enough. Firstly, more granular temperature data could be added. In addition, we are only utilizing the highs and lows of the day; weather data is more nuanced than two data points, hourly or more accurate temperatures could lead to more specific inferences--stronger correlations. 
2. Single source of info. We are only investigating the crime data with regards to the temperature localized on a single weather station. The temperature should be used as a reference and not as an exact indicator of the temperature at the time of the dispatch.
3. Single point of failure. By only using one weather station's reported temperature data, this approach could lead to holes in our data. For example, certain days in which the station did not report the weather, for various reasons, our dataset had no results (missing values) for certain days.

### A selection of data for continued analysis with justifications:
---

#### Crime Data Trimming
There are approximately 830,000 or so crime reports that show up as neither misdemeanor nor felonies. "All other offenses" and "all other assaults" are both over-represented in our dataset. Due to their very unspecific nature of being labeled as "other," we can not assign a boolean of misdemeanor or felony. In practice, the Philadelphia Police Department only has so many bins in which to categorize crimes. In our analysis, it is important that when we are looking at felony vs misdemeanor, we only look at them in comparison to each other, and not out out of the whole.

Our group looked at various sources for methodologies of categorizing crime. The most coherent--but not necessarily the most relevant--categorization is that of felony vs misdemeanor. In addition we also classified crimes based on the presence of violence.

The classification of a crime as a felony or a misdemeanor is that of judicial matter; some crimes can be either, when presented in court by a district attorney. Our data is simply crime incident reporting--not conviction data. In example, from our dataset's perspective, a theft is simply a theft. However, when presented in a judicial setting, a theft can be classified as a misdemeanor (on the low end) to a felony (on the high end) depending on the valuation of what was stolen. For the sake of data coherency and integrity, we decided to take all variable crimes (misdemeanor or felony, depending on certain criteria) and exclude them from investigations that are specifically about "misdemeanor versus felony"-reporting.

The source of classification regarding "misdemeanor versus felony" comes from the Pennsylvania State Legislature, primarily sourced from [Title 18](https://www.legis.state.pa.us/WU01/LI/LI/CT/HTM/18/18.HTM)  which defines various crimes and the grading in which they can be prosecuted.

The source of classification on violent crimes is from the FBI Violent Crime [Reporting System](https://ucr.fbi.gov/crime-in-the-u.s/2018/crime-in-the-u.s.-2018/topic-pages/violent-crime). We co-opted to use their classification system, as the crime typology falls in line with how the State of Pennsylvania categorizes their crimes.

## Daily-level dataset

```python
data = dispatch_level_df.copy()
data['dispatch_date'] = pd.to_datetime(data['dispatch_date'])

# aggregate violent offenses by date
df1 = data.copy()
df1 = df1.groupby(['dispatch_date', 'violent_offense']).size().reset_index(name='violent_offense_count')
df1 = df1[df1['violent_offense'] == "yes"]
df1 = df1.drop(columns=['violent_offense'])
df1 = df1.reset_index(drop=True)

# aggregate misdemenors and felonies by date
df2 = data.copy()
df2 = (df2
        .groupby(['dispatch_date', 'mis_felony'])
        .agg(mis_felony_count=('mis_felony', 'count'))
        .reset_index()
        .pivot(index='dispatch_date', columns='mis_felony', values='mis_felony_count')
        .reset_index()
        .rename(columns={'felony': 'felony_count', 'misdemeanor': 'misdemeanor_count'}))

xdata = (data.loc[:, ['dispatch_date', 'year', 'year_group', 'dispatch_wday',
                      'dispatch_month', 'dispatch_quarter', 'tmax',
                      'tmin']]
         .drop_duplicates(subset=['dispatch_date'], keep='first'))

daily_aggregated_df = (df1.merge(df2, on='dispatch_date', how='left')
                       .merge(xdata, on='dispatch_date', how='left'))

daily_aggregated_df.head()
```
