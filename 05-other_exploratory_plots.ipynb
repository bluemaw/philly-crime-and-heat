```python
import os
import re
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style('whitegrid')
```

```python
daily_df = pd.read_csv('data/processed/daily_aggregated_df.csv')
```

```python
# Visualizing the relationship between violent offense count and temperature on a daily basis using a seaborn distribution plot
# The distribution plots are categorized via columns by dispatch quarter (AKA quarter/season)
# The scale of the violent_offense_counts is transformed using a natural log scale (base e) for better visualization/interpretation
ax = sns.displot(data=daily_df, x='tmax', y='violent_offense_count', col='dispatch_quarter', alpha=0.5, height=6, aspect=1)

ax.set(yscale='log')

plt.show()
```

```python
# The same visualization as above but with more discrete bins of data points for visualization purposes
ax = sns.displot(data=daily_df, x='tmax', y='violent_offense_count', col='dispatch_quarter', alpha=0.5, height=6, aspect=1, discrete=(True,True))

ax.set(yscale='log')

plt.show()
```

```python
# Visualizing the relationship between felony count and temperature on a daily basis using a seaborn distribution plot
# The distribution plots are categorized via columns by dispatch quarter (AKA quarter/season)
# The scale of the felony counts is transformed using a natural log scale (base e) for better visualization/interpretation
ax = sns.displot(data=daily_df, x='tmax', y='felony_count', col='dispatch_quarter', alpha=0.5, height=6, aspect=1)

ax.set(yscale='log')

plt.show()
```

```python
# Visualizing the relationship between misdemeanor counts and temperature on a daily basis using a seaborn distribution plot
# The distribution plots are categorized via columns by dispatch quarter (AKA quarter/season)
# The scale of the misdemeanor counts is transformed using a natural log scale (base e) for better visualization/interpretation
ax = sns.displot(data=daily_df, x='tmax', y='misdemeanor_count', col='dispatch_quarter', alpha=0.5, height=6, aspect=1)

ax.set(yscale='log')

plt.show()
```

```python
# Visualizing the relationship between violent offense counts and tmax using a seaborn relplot based on a hue of years
# Found that binning the data by years didn't help very much in terms of wanting to find further insight
sns.relplot(data=daily_df, y='violent_offense_count', x='tmax', hue='year')

plt.show()
```

```python
# After looking at years, I decided to bin the data by quarters instead
# Visualizing the relationship between violent offense counts and tmax using a seaborn relplot
sns.relplot(data=daily_df, y='violent_offense_count', x='tmax', hue='dispatch_quarter', alpha=0.5, kind='scatter',
            height=6, aspect=1)

plt.show()
```

```python
# Visualizing the relationship between violent offense counts and tmax using a seaborn relplot categorizing the columns and hue by quarter
# Found that this visualization is interesting and pretty, but isn't as helpful as the previous visualization in terms of confirming inferences from data
ax = sns.relplot(data=daily_df, x='violent_offense_count', y='tmax', hue='dispatch_quarter', col='dispatch_quarter', alpha=0.5, kind='scatter', height=6, aspect=1)

ax.set(xscale='log')

plt.show()
```

```python

# Visualizing the relationship between violent offense counts (using a natural log transformation) and tmax
# using a seaborn relplot overlaying hues based on quarters (warmer hues are quarters 2 and 3, where as quarters 1 and 2 are cooler, inherently)
ax = sns.relplot(data=daily_df, y='violent_offense_count', x='tmax', hue='dispatch_quarter', alpha=0.5,
                 kind='scatter', height=6, aspect=1)

ax.set(xscale='log')

plt.show()
```
